```python
initial_money = 303818495
stock_price = 235.33

money = initial_money
half_money = initial_money / 2
max_shares = initial_money // stock_price
half_shares = max_shares // 2

while money >= stock_price:
    
    money -= stock_price
    shares += 1
    
    print("1주 구매했습니다.")
    
    if shares >= half_shares:
        print("매수 가능한 횟수 범위를 넘었습니다!")
        break
        
    if money <= half_money:
        print("매수 가능한 금액 범위를 넘었습니다!")
        break

print("총 구매한 주식 수: ", shares)
print("총 소비한 돈: ", initial_money-money)
print("남은 돈: ", money)
    
```
